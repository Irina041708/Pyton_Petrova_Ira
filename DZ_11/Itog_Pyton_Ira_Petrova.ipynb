Дана функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0


```python
import sympy
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols
from sympy import solveset
from sympy.abc import x 
from sympy import sin, cos, pi
import pylab
```

1. Определим корни уравнения (ans - это корни уравнений)

```python
x = sympy.symbols('x')
```

```python
f = -12*(x**4)*sin(cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
```

```python
ans = solveset(f)
```

```python
ans
```

ans - это корни уравнений, которых в данном примере может быть бесконечное множество комплексных чисел.Чтобы найти определенное количество корней, необходимо,чтобы в постановке задачи изначально указали промежуток, в котором хотели бы найти корни уравнения.Пример на практике разбирали гораздо проще и при решении этой задачи , нам не объяснили какие инструменты применяют при решении данной задачи. Поэтому моё решение основано с помощью графика(приближенно).

```python
x = np.arange(-20, -15, 0.01)
f = -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30

plt.plot(x, f,'--g')
plt.grid(True,color='b')


```

Возьмеме промежуток от [-20;-15]

Функция f возрастает на промежутке ~(-19;-16)

Функция f убывает на промежутках [-20;-19) и (-16;-15]

При x = -19 - минимум(вершина)

При x = -16 - максимум(вершина)

При  (-17,4)< x <= (-15) функция f больше нуля.

При  (-20)<= x < (-17,4) функция f меньше нуля.

При x = 17,4 функция f = 0



```python
from sympy import *
x = Symbol('x')
print(solveset(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, x, domain=S.Reals))
```
